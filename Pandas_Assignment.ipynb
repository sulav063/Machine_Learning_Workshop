````python
# Read csv file world_population.csv
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pandas/world_population.csv")
df
````

````python
# Display the list of countries having a rank<= 10
df1 = df[df["Rank"] <= 10]
df1
````

````python
# Create a list of specific country ['Bangladesh', 'Brazil'] and return df which falls under the specific list only
country = ["Bangladesh", "Brazil"]
ba=df[df["Country"].isin(country)]
ba
````

````python
# Apply filter to the Country which has United in its name
united_countries = df[df['Country'].str.contains('United')]
united_countries
````

````python
# Create a dataframe df2 and give Country as its index
df2 = df.copy()
df2 = df2.set_index('Country')
df2
````

````python
# Apply filter Continent and CCA3 in df2 (need only two columns)
filtered_df2 = df2[['Continent', 'CCA3']]
filtered_df2
````

````python
# Apply filter to see details of Zimbabwe entire row
zimbabwe_row = df[df['Country'] == 'Zimbabwe']
zimbabwe_row
````

````python
# Table 3
df1 = df.sort_values(by=['Continent', 'Country'], ascending=[False,True])
df1
````
