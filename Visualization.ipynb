# Why Choose Matplotlib and Seaborn?

Both libraries are among the most popular choices for data visualization in Python. Hereâ€™s why:

## Basic of Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
```

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]
colors = ['skyblue', 'lightcoral', 'lightgreen']
plt.bar(categories, values, color=colors)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Visualizations with Seaborn

### Histogram and KDE plot

```python
import seaborn as sns
```

```python
data = np.random.randn(1000)
sns.histplot(data, kde=True, color='Green')
plt.title("Histogram with KDE")
plt.show()
```

### Heatmap

```python
matrix = np.random.rand(5, 5)
sns.heatmap(matrix, annot=True, cmap="prism")
plt.title("Heatmap Example")
plt.show()
```

### Pair Plot

```python
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pair Plot Example", y=1.02)
plt.show()
```
